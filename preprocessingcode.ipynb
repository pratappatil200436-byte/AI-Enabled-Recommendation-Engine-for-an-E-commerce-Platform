```python
import pandas as pd
import numpy as np
```

```python
orders = pd.read_csv("/content/olist_orders_dataset.csv")
order_items = pd.read_csv("/content/olist_order_items_dataset.csv")
customers = pd.read_csv("/content/olist_customers_dataset.csv")
products = pd.read_csv("/content/olist_products_dataset.csv")
```

```python
# Merge order items with orders
data = pd.merge(
    order_items,
    orders[['order_id', 'customer_id', 'order_status']],
    on='order_id',
    how='inner'
)

# Keep only delivered orders
data = data[data['order_status'] == 'delivered']

# Merge products
data = pd.merge(
    data,
    products[['product_id', 'product_category_name']],
    on='product_id',
    how='left'
)

# Merge customers
data = pd.merge(
    data,
    customers[['customer_id', 'customer_unique_id']],
    on='customer_id',
    how='left'
)
df=pd.DataFrame(data)
df.shape
df.head()
```

```python
transactions = data[['customer_unique_id', 'product_id', 'product_category_name']]

transactions.rename(columns={
    'customer_unique_id': 'user_id',
    'product_id': 'item_id',
    'product_category_name': 'category'
}, inplace=True)
```

```python
print("\nNull values count:")
print(transactions.isnull().sum())

print("\nDuplicate rows count:")
print(transactions.duplicated().sum())

print("\nTotal rows:")
print(transactions.shape[0])
```

```python
# Remove missing values
transactions.dropna(inplace=True)
# Remove duplicate interactions
transactions.drop_duplicates(inplace=True)
# Reset index
transactions.reset_index(drop=True, inplace=True)
t=pd.DataFrame(transactions)
t.head()
```

```python
print("\nNull values count:")
print(transactions.isnull().sum())

print("\nDuplicate rows count:")
print(transactions.duplicated().sum())

print("\nTotal rows:")
print(transactions.shape[0])
```

```python
transactions['interaction'] = 1
```

```python
# Keep users with at least 2 interactions
user_counts = transactions['user_id'].value_counts()
valid_users = user_counts[user_counts >= 2].index
transactions = transactions[transactions['user_id'].isin(valid_users)]

# Keep items with at least 2 interactions
item_counts = transactions['item_id'].value_counts()
valid_items = item_counts[item_counts >= 2].index
transactions = transactions[transactions['item_id'].isin(valid_items)]
```

```python
transactions.to_csv("cleaned_ecommerce_dataset.csv", index=False)

print("Final cleaned dataset shape:", transactions.shape)
```

```python
from scipy.sparse import csr_matrix

# Encode users and items
user_codes = transactions['user_id'].astype('category').cat.codes
item_codes = transactions['item_id'].astype('category').cat.codes

# Create sparse matrix
user_item_sparse = csr_matrix(
    (transactions['interaction'], (user_codes, item_codes))
)

print("User–Item Sparse Matrix Shape:", user_item_sparse.shape)
```
